```python
def mkplot(col1, col2, col3, eqname, regression):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(projection="3d")
    labelsize = 20
    ax.set_zlim(-2,2)
    #ax.set_ylim(-1,1)
    #ax.set_xlim(-1,1)
    ax.set_xlabel(col1, fontsize= labelsize)
    ax.set_ylabel(col2, fontsize= labelsize)
    ax.set_zlabel(col3, fontsize= labelsize)
    ax.set_title(eqname)
#     ax.set_title(f"Trained Agent vs AI Feynman Regression {eqname}", fontsize=10)
    ax.plot(nn_data[col1], nn_data[col2], nn_data[col3], label="Trained Agent")#,c=nn_data[col3], cmap="viridis") 
    ax.plot(nn_data[col1], nn_data[col2], regression, color="red", label="AI Feynman Regression")
    ax.legend()


    
```

```python
# Get function from string
def get_fn(eqn):
    eqn.replace("pi", "3.14159")
 

    fn = sympify(eqn, evaluate=False)
    #fn.subs(x2, 3)
    #fn.subs(x1, 100)

    # eqn.replace("x1", "3")
    # eqn.replace("x0", "3")
    return fn
```
